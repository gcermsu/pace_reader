# PACE READ EXAMPLE

### Objective:
Convert a PACE satellite .nc image file into a georeferenced .tiff file.

### Dataset Overview:
The PACE satellite image dataset is organized by band names such as rhot_blue, rhot_red, and rhot_SWIR. Each band contains a series of enumerated subsets. The goal is to convert either the entire subset of images into a single .GeoTIFF file or extract and save a specific band of interest as a .GeoTIFF.

### Requirements:
The code should be capable of the following:

- **Read the .nc file:** Import and handle NetCDF files containing PACE satellite image data.

- **Select the Band of Interest:** Allow users to choose a specific band or subset of bands for processing.

- **Visualize the Georeferenced Image Band:** Display the selected image band with proper georeferencing to ensure accurate spatial representation.

- **Export as GeoTIFF:** Convert and export the selected band or the entire subset of bands into a georeferenced .GeoTIFF file, suitable for further analysis or integration with other GIS tools.

```python
import os
import xarray as xr

from PACEReadPack import read_PACE
```

Define the directory of the .nc file, the desired band ('blue', 'red', or 'SWIR'), and the band number (optional)

```python
nc_file = r'C:\Users\tml411\Documents\PACE_geotransformTIFF\PACE_geotransformTIFF\PACEReadPackage\PACE_OCI.20240601T182541.L1B.V2.nc'
band_name = 'blue'
band_number = None # If None, the entire band will be reprojected and saved as a unique .tiff file. Otherwise, only the desired band will be reprojected and saved.
```

Initially we will read the .nc file using Xarray and plot the desired band reprojected, just for visualization purposes.

```python
obs_dataset = xr.open_dataset(nc_file, group='observation_data')
obs_dataset = obs_dataset.rename({'number_of_scans': 'number_of_lines', 'ccd_pixels': 'pixels_per_line'})
rrs = obs_dataset['rhot_' + band_name].transpose('number_of_lines', 'pixels_per_line', band_name + '_bands')

# Read geolocation data to reproject the image band in terms of latitude and longitude
dataset_geo = xr.open_dataset(nc_file, group='geolocation_data')
dataset_geo = dataset_geo.set_coords(('longitude', 'latitude'))
dataset_geo = dataset_geo.rename({'number_of_scans': 'number_of_lines','ccd_pixels': 'pixels_per_line'})

dataset_merged = xr.merge((rrs, dataset_geo.coords))

dataset_merged = rrs.assign_coords({'y': dataset_geo.coords['latitude'], 'x': dataset_geo.coords['longitude']})
dataset_merged = dataset_merged.drop_vars('latitude').drop_vars('longitude').rename({'number_of_lines': 'y','pixels_per_line': "x"})
dataset_merged = dataset_merged.transpose(band_name + '_bands', 'y', 'x')

if band_number == None: data_geo = dataset_merged.sel({band_name + '_bands': 1})
else: data_geo = dataset_merged.sel({band_name + '_bands': band_number})

plot = data_geo.plot(x='x', y='y', cmap='viridis', vmin=0)
```

Now we can read the .nc file and convert it to a reprojected GeoTiff.

```python
read_PACE.read_PACE(nc_file, band_name, band_number)
```
